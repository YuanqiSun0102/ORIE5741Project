### Some distribution of the data feature

```python
columns = np.array(df.columns)
plt.figure(figsize=(18, 12))

for i, column in enumerate(columns):
    plt.subplot(6, 3, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.tight_layout()

plt.show()
```

## Categorical

```python
fg = sns.catplot(x="Revenue", y="PageValues", data=df, kind="box", dodge=True)
```

```python
cate_columns = ['VisitorType', 'Browser', 'Region', 'OperatingSystems', 'TrafficType', 'Weekend']
```

```python
num_plots = len(cate_columns)
num_rows = (num_plots + 2) // 3 

plt.figure(figsize=(15, 5 * num_rows)) 

for i, col in enumerate(cate_columns):
    plt.subplot(num_rows, 3, i+1) 
    sns.violinplot(x=col, y="Revenue", data=df, dodge=True)
    plt.title(f'{col}')

plt.tight_layout() 
plt.show()
```

## Correlation

```python
df['Month'].value_counts()
```

### Turn string type to int/float

```python
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

df['Month'] = df['Month'].map(month_mapping)
```

### Encoding the visitor_type

```python
df_encoded = pd.get_dummies(df, columns=['VisitorType', 'Browser', 'Region', 'OperatingSystems','TrafficType'])
df_encoded
```

```python
df_encoded.describe()
```

```python
corr = df_encoded.corr()
corr
```

```python
from scipy import stats
ax = plt.subplots(figsize=(50, 40))
ax = sns.heatmap(corr, vmax=.8, square=True, annot=True,cmap="RdYlGn")
```
